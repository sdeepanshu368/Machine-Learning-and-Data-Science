```python
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
plt.plot([1,2,3],[2,5,7])
plt.title('Title here')
plt.ylabel('This is Y')
plt.xlabel('This is X')
plt.show()
```

```python
x = [1,5,4]
y = [23,45,89]
x1 = [3,7,3]
y1 = [5,8,9]
plt.plot(x,y)
plt.plot(x1,y1)
plt.show()
```

```python
from matplotlib import style
style.use('ggplot')
x = [1,5,4]
y = [23,45,89]
x1 = [3,7,3]
y1 = [5,8,9]
plt.plot(x,y)
plt.plot(x1,y1)
plt.show()
```

```python
x = [1,5,4]
y = [23,45,89]
x1 = [3,7,3]
y1 = [5,8,9]
plt.plot(x,y, label='First', linewidth=5)
plt.plot(x1,y1, label='Second')
plt.legend()
plt.show()
```

```python
x = [1,5,4]
y = [23,45,89]
x1 = [3,7,3]
y1 = [5,8,9]
plt.scatter(x,y, label='First', linewidth=5)
plt.scatter(x1,y1, label='Second')
plt.legend()
plt.show()
```
